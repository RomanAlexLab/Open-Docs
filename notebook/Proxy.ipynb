```python
import os
import httpx
import openai
from openai import OpenAI
from openai import AsyncOpenAI
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_openai import OpenAIEmbeddings
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
```

```python
load_dotenv()
```

```python
# Загружаем ключ API OpenAI
openai_key = os.getenv('OPENAI_API_KEY')
os.environ["OPENAI_API_KEY"] = openai_key
openai.api_key = openai_key
```

```python
MODEL_GPT = "gpt-3.5-turbo"
MODEL_GPT_TEMP = 0.7
```

# API

**Синхронный**

```python
proxies = {
    'http://': 'http://Rfghtr:wewere@145.319.51.82:66456',
    'https://': 'http://Rfghtr:wewere@145.319.51.82:66456'
}

transport = httpx.HTTPTransport(local_address="0.0.0.0")

session = httpx.Client(proxies=proxies, transport=transport)
```

```python
client = OpenAI(http_client=session)
```

```python
completion = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "system", "content": "Ты бот"},
    {"role": "user", "content": "Привет бот"},
  ]
)

print(completion.choices[0].message)
```

**Асинхронный**

```python
aclient = AsyncOpenAI(http_client=httpx.AsyncClient(proxies="http://Rfghtr:wewere@145.319.51.82:66456",
                                                   transport=httpx.AsyncHTTPTransport(local_address="0.0.0.0")))
```

```python
completion = await aclient.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "system", "content": "Ты бот"},
    {"role": "user", "content": "Привет бот"},
  ]
)

print(completion.choices[0].message)
```

# LangChain

**Синхронный**

```python
proxies = {
    'http://': 'http://Rfghtr:wewere@145.319.51.82:66456',
    'https://': 'http://Rfghtr:wewere@145.319.51.82:66456'
}

transport = httpx.HTTPTransport(local_address="0.0.0.0")

session = httpx.Client(proxies=proxies, transport=transport)
```

```python
# Создаём объект OpenAIEmbeddings
embeddings = OpenAIEmbeddings(http_client=session)
```

```python
chat = ChatOpenAI(model=MODEL_GPT, temperature=MODEL_GPT_TEMP, http_client=session)
```

```python
# Формируем сообщение
messages = [
    SystemMessage(content='Ты консультант'),
    HumanMessage(content='Привет бот')
]
```

```python
chat.invoke(messages)
```

**Асинхронный**

```python
proxies = {
    'http://': 'http://Rfghtr:wewere@145.319.51.82:66456',
    'https://': 'http://Rfghtr:wewere@145.319.51.82:66456'
}

transport = httpx.AsyncHTTPTransport(local_address="0.0.0.0")

async_client = httpx.AsyncClient(proxies=proxies, transport=transport)
```

```python
# Создаём объект OpenAIEmbeddings
embeddings = OpenAIEmbeddings(http_async_client=async_client)
```

```python
chat = ChatOpenAI(model=MODEL_GPT, temperature=MODEL_GPT_TEMP, http_async_client=async_client)
```

```python
# Формируем сообщение
messages = [
    SystemMessage(content='Ты консультант'),
    HumanMessage(content='Привет бот')
]
```

```python
await chat.ainvoke(messages)
```
